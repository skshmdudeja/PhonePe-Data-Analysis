```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['font.size'] = 11
```

```python
FILE = "Phonepe-Final-Dataset.xlsx"   # <-- change this to your file name

users        = pd.read_excel(FILE, sheet_name="All_Users")
transactions = pd.read_excel(FILE, sheet_name="All_Transactions")
recharge     = pd.read_excel(FILE, sheet_name="Recharge_Bills")
transfer     = pd.read_excel(FILE, sheet_name="Money_Transfer")
loans        = pd.read_excel(FILE, sheet_name="Loans")
insurance    = pd.read_excel(FILE, sheet_name="Insurance")

print(" All tables loaded successfully!\n")
```

```python
tables = {
    "All_Users": users,
    "All_Transactions": transactions,
    "Recharge_Bills": recharge,
    "Money_Transfer": transfer,
    "Loans": loans,
    "Insurance": insurance
}

print("=" * 50)
print("TABLE SHAPES (rows x columns)")
print("=" * 50)
for name, df in tables.items():
    print(f"{name:20s} → {df.shape[0]:,} rows, {df.shape[1]} columns")

print("\n" + "=" * 50)
print("MISSING VALUES PER TABLE")
print("=" * 50)
for name, df in tables.items():
    missing = df.isnull().sum().sum()
    print(f"{name:20s} → {missing} missing values")
```

```python
print("\n" + "=" * 50)
print("USER ANALYSIS")
print("=" * 50)

# Convert Join_Date to datetime
users['Join_Date'] = pd.to_datetime(users['Join_Date'])
users['Join_Month'] = users['Join_Date'].dt.to_period('M')
users['Join_Year']  = users['Join_Date'].dt.year

# Age distribution
print(f"Average User Age : {users['Age'].mean():.1f} years")
print(f"Youngest User    : {users['Age'].min()} years")
print(f"Oldest User      : {users['Age'].max()} years")

# Age groups
bins   = [0, 25, 35, 45, 60, 100]
labels = ['18-25', '26-35', '36-45', '46-60', '60+']
users['Age_Group'] = pd.cut(users['Age'], bins=bins, labels=labels)
```

```python
print("\n" + "=" * 50)
print("ALL TRANSACTIONS ANALYSIS")
print("=" * 50)

transactions['Date'] = pd.to_datetime(transactions['Date'])
transactions['Month'] = transactions['Date'].dt.to_period('M')

print(f"Total Transactions : {len(transactions):,}")
print(f"Total Revenue      : ${transactions['Amount'].sum():,.2f}")
print(f"Avg Transaction    : ${transactions['Amount'].mean():,.2f}")

# Success vs Failure rate
status_counts = transactions['Payment_Status'].value_counts()
print(f"\nPayment Status Breakdown:")
print(status_counts)
success_rate = (status_counts.get('Successful', 0) / len(transactions)) * 100
print(f" Success Rate: {success_rate:.1f}%")
```

```python
print("\n" + "=" * 50)
print("RECHARGE & BILLS ANALYSIS")
print("=" * 50)

recharge['Date'] = pd.to_datetime(recharge['Date'])

print(f"Total Recharge Transactions : {len(recharge):,}")
print(f"Total Recharge Revenue      : ${recharge['Amount'].sum():,.2f}")
print(f"\nRecharge Type Breakdown:")
print(recharge['Recharge_Type'].value_counts())

```

```python
print("\n" + "=" * 50)
print("MONEY TRANSFER ANALYSIS")
print("=" * 50)

transfer['Date'] = pd.to_datetime(transfer['Date'])

print(f"Total Transfer Transactions : {len(transfer):,}")
print(f"Total Amount Transferred    : ${transfer['Amount'].sum():,.2f}")
print(f"Average Transfer Amount     : ${transfer['Amount'].mean():,.2f}")
print(f"\nTransfer Type Breakdown:")
print(transfer['Transfer_Type'].value_counts())
```

```python
print("\n" + "=" * 50)
print("LOANS ANALYSIS")
print("=" * 50)

loans['Date'] = pd.to_datetime(loans['Date'])

print(f"Total Loan Transactions : {len(loans):,}")
print(f"Total Loan Amount       : ${loans['Loan_Amount'].sum():,.2f}")
print(f"Average Loan Amount     : ${loans['Loan_Amount'].mean():,.2f}")
print(f"Max Loan Amount         : ${loans['Loan_Amount'].max():,.2f}")
print(f"\nLoan Type Breakdown:")
print(loans['Loan_Type'].value_counts())
```

```python
print("\n" + "=" * 50)
print("INSURANCE ANALYSIS")
print("=" * 50)

insurance['Date'] = pd.to_datetime(insurance['Date'])

print(f"Total Insurance Transactions : {len(insurance):,}")
print(f"Total Premium Collected      : ${insurance['Premium'].sum():,.2f}")
print(f"Average Premium              : ${insurance['Premium'].mean():,.2f}")
print(f"\nInsurance Type Breakdown:")
print(insurance['Insurance_Type'].value_counts())
```

```python
print("\n" + "=" * 50)
print("FINAL BUSINESS SUMMARY")
print("=" * 50)

total_users    = len(users)
total_txn      = len(transactions)
total_revenue  = transactions['Amount'].sum()
top_service    = transactions.groupby('Service')['Amount'].sum().idxmax()
top_loan       = loans.groupby('Loan_Type')['Loan_Amount'].sum().idxmax()
top_insurance  = insurance.groupby('Insurance_Type')['Premium'].sum().idxmax()
top_recharge   = recharge.groupby('Recharge_Type')['Amount'].sum().idxmax()
top_transfer   = transfer.groupby('Transfer_Type')['Amount'].sum().idxmax()

print(f"👤 Total Users           : {total_users:,}")
print(f"💳 Total Transactions    : {total_txn:,}")
print(f"💰 Total Revenue         : ₹{total_revenue:,.2f}")
print(f"🏆 Top Service           : {top_service}")
print(f"🏦 Top Loan Type         : {top_loan}")
print(f"🛡️  Top Insurance Type   : {top_insurance}")
print(f"📱 Top Recharge Type     : {top_recharge}")
print(f"💸 Top Transfer Type     : {top_transfer}")
print(f"✅ Transaction Success % : {success_rate:.1f}%")
```
